# More Regex

```python
url = "https://www.518.com.tw/job-index.html?ad=Android&aa=&ab=&ac=&am=&i="
```

```python
import re

import requests

resp = requests.get(url)
```

```python
resp.text
```

```python
re.findall('''(https://www.518.com.tw/.+html)\?kw=Android&pi=1''', resp.text)
```

```python
s = '''https://www.518.com.tw/技術部_Android軟體工程師-台北市-松山區-job-1196612.html.html.html.html.html.html.html.htmlaosdgpasdfaisudfias'''
```

```python
# +預設是找到右邊的最後一個才停(Greedy)
# +? 可以把+變成Lazy，也就是找到第一個match就停
re.findall('https://www.518.com.tw/.+?html', s)
```

```python
re.findall('https?://www.518.com.tw/.+?html', s)
```
