# 川普就職演說詞頻分析

```python
with open('./datasets/trump_speech.txt') as f:
    speech = f.read()
```

```python
with open('./datasets/eng_stop_words.txt') as f:
    stop_words = f.read().split('\n')
```

```python
stop_words
```

```python
speech[:100]
```

```python
from collections import Counter
wc = Counter() # Dictionary like object

for word_raw in speech.split():
    word = word_raw.replace(',', '').replace('.', '').lower()
    if word not in stop_words:
        if word in wc:
            wc[word] += 1
        else:
            wc[word] = 1
```

```python
wc.most_common()
```

```python
# [{'1': 'record 1'},
#  {'2': 'record 2'}]
gym_data = []
with open('./datasets/gym.csv') as f:
    for idx, row in enumerate(f): # enumerate -> 得到for loop的index
        if idx == 0:
            header = row.strip().split(",")
        else:
            # zip(['key1', 'key2], [1,2]) -> (key1, 1), (key2, 2)
            # dictionary comprehension..
            row_dict = {k:v for k,v in zip(header, row.strip().split(","))}
            gym_data.append(row_dict)
        print("NO %s - %s" % (idx, row.strip()))
```
