```python
import re

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res = requests.post("http://coolpc.com.tw/m/m-list.php", data={"G":7})
```

```python
res
```

```python
soup = BeautifulSoup(res.text, 'lxml')
```

```python
ssds = soup.select("span.Q > table")[1].find_all(attrs={"onclick": 'A(this)'})
```

```python
data = [row.text for row in ssds]
```

```python
data
```

```python
data[4]
```

```python
re.findall('(\d+)(TB|G)', data[4])
```

```python
re.findall('\$(\d+)', data[4])
```

```python
re.findall('寫:(\d+)M', data[4])
```

```python
re.findall('[T|M]LC', data[4])
```

```python
result = []
for row in data:
    ssd = {}
    try:
        ssd['name'] = row.split("/")[0]
        cap = re.findall('(\d+)(TB|G)', row)[0]
        if cap[1] == 'G':
            ssd['capacity'] = int(cap[0])
        else:
            ssd['capacity'] = int(cap[0]) * 1000
        ic = re.findall('[T|M]LC', row)
        ssd['ic'] = ic[0] if ic else ' ' 
        ssd['read'] = int(re.findall('讀:(\d+)M', row)[0])
        ssd['write'] = int(re.findall('寫:(\d+)M', row)[0])
        ssd['price'] = int(re.findall('\$(\d+)', row)[0])
        ssd['warranty'] = re.findall('(.)年', row)[0]
        result.append(ssd)
    except:
        print(row)
result
```

```python
df = pd.DataFrame(result)
```

```python
df.head()
```

```python
df['cp'] = df.price / df.capacity
```

```python
df.head()
```

```python
df.sort_values('cp')[(df.price < 3000) & (df.ic == 'MLC') & (df.warranty == '五')]
```
